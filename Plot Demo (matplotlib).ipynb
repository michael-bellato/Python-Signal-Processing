## Wavelet

```python
sr = 44100                             
dt = 1/sr                             
dur = 2                                
time = np.arange(-dur/2,(dur/2)-dt,dt)      
```

```python
freq = 15                           
amp = 1                            
wcy = 6                    
```

```python
sinewave = np.cos(time*2*pi*freq);                                   
plt.plot(time,sinewave)
```

```python
#Construct gaussian window
s = wcy/(2*pi*freq) # standard deviation of sine wave
gaussw = amp*np.exp(-(time)**2/(2*(s**2)))                         
plt.plot(time,gaussw)
```

```python
#Convolve sine wave with window to create Morelet Wavelet
mwlt = sinewave*gaussw
plt.plot(time,mwlt)  
```

```python
s = 16/(2*pi*freq) # standard deviation of sine wave
gaussw = amp*np.exp(-(time)**2/(2*(s**2)))                         
plt.plot(time,gaussw)
```

```python
def get_b(x, y):
    numerator = np.sum((x*y) - (y.mean()*x))
    denominator =  np.sum((x**2) - (x.mean()*x))
    return numerator / denominator
def get_a(x, y):
    return np.mean(y) - get_b(x, y)*np.mean(x)
```
